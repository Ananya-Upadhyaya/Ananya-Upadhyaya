```python
# Group the people who obtained marks above or below 70

above_70 =exam_results['Scores'] > 70
above_70

below_70 =exam_results['Scores'] < 70
below_70
```

```python
# the mean scores of the students who scored above or below 70

exam_results.groupby(above_70).mean()

exam_results.groupby(below_70).mean()
```

```python
# Merging and Joining

contacts = pd.DataFrame(
    {
     'Name' :['Aisha','Charlie'] ,
     'Tel'  :['+91 123 456 78 89', '+91 143 234 54 67'] 
    }
)

contacts
```

```python
 # defaults to inner join
exam_results.merge(contacts, left_on='Name', right_on='Name')  

 # defaults to outer join

 exam_results.merge(contacts, left_on='Name', right_on='Name', how = 'outer')  
```
