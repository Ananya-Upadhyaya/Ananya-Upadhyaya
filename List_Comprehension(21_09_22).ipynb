```python
# 1. Find all of the numbers from 1-1000 that are divisible by 8. 

List = [n for n in range(1,1001) if n % 8 == 0] 
print(List)
```

```python
#2. Find all of the numbers from 1-1000 that have a 6 in them. 

num = [n for n in range(1,1001) if '6' in str(n)]
print(num)
```

```python
#3. Count the number of spaces in the following string. String:  “Practice problems to drill list comprehension in your head” 

a=input("Enter the string paragraph:")
count=0
for i in a:
  if  i==" ":
       count=count+1
       print("Number of spaces in a string:",count)
```

```python
#4. Remove all the vowels in the above string. 

string = input("Enter the string: ")

vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
result = ""

for char in string:
    if char not in vowels:
        result = result + char

print("\nAfter removing Vowels: ", result)
```

```python
#5. Find all the words in the above string that are less than 5 letters. 

Str = 'Practice problems to drill list comprehension in your head'

examine = Str.split()

result = [word for word in examine if len(word) <=5]
print(result)
```

```python
# Challenge
#1. use dictionary comprehension to count the length of each word in  the sentence (refer to the above string)

def length_of_words(s):
    return {word:len(word) for word in s.split()}
print(length_of_words("Practice problems to drill list comprehension in your head"))
```

```python
# 2. use a nested list comprehension to find all of the numbers from 1-1000 that are divisible by any single digit besides 1 (2-9) 

result = [number for number in range(1,1001) if True in [True for x in range(2,10) if number % x == 0]]
print(result)
```

```python
# 3. For all the numbers 1-1000, use a nested list/dictionary comprehension to find the highest single digit any of the numbers is divisible by. 

highest_digit = {num:max(div for div in range(2,10) if num % div == 2) for num in range(1,1001) if [div for div in range(2,10) if num % div == 2]}
print(highest_digit) 
```
